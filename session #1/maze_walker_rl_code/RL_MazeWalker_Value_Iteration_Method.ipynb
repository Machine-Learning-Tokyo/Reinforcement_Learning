### VALUE ITERATION METHOD ###

Value iteration is a **GREEDY** algorithm. Unlike policy iteration where we had 2 different steps for policy evaluation and improvement, in this method policy improvement is embedded withtin the policy evaluation step.

- Policy Evaluation Step
        : Policy Improvement Step (Embedded in each Evaluation Step)


- Base Equation

  $V(s) = \sum_a\pi(a|s)(\sum_\hat{s}\sum_r [p(\hat{s}, r | s, a)(r + \gamma*v(\hat{s})])$
  
At the base level, we continue to use the above equation. The only change is the way in which the policy is updated.


<h4 style="color:blue">POLICY ITERATION ALGORITHM (What we saw earlier)</h4>

```
- Build an arbitary policy
- Associate some random state values to each state
- continue_flag = True
- while continue_flag
  - for every state                                                --------------
    - update state values using current policy and state_value function         | ------> POLICY EVALUATION
      equation [v(s)]                                                           |
    - continue till state values converge                         ---------------
  - for every state                                               ---------------
    - for every action                                                          | ------> POLICY IMPROVEMENT
      - using current policy, check the update in state value                   |
        using value function equation [v(s)]                                    |
    - select the action which leads to maximum increase in        ---------------
      state value and update the policy for this state
  if old_policy == new_policy
    - continue_flag = False
```

<h4 style="color:red">VALUE ITERATION ALGORITHM (A modification over policy iteration)</h4>

```
 - Buils an arbitary policy
 - Associate some random state values to each state
 - continue_flag = True
 - while continue_flag
   - for every state                                                     ----------        
     - for every action                                                           |------> GREEDY APPROACH   
       - using current policy, check the update in state value using value        |
         function equation                                                        |
     - select the action which leads to maximum increase in state value           |------> POLICY IMPROVEMENT
     - get next_state and reward based on chosen action                           |        EMBEDDED WITHIN
     - use next_state and reward in state value function equation [v(s)]          |        POLICY EVALUATION
       to update current state value.                                             |
     - update the policy based on chosen action                          ----------
   - if old_policy == new_policy
     - continue_flag = False
```

<h3> NOTE </h3>
- We can see that VALUE ITERATION method was able to learn much faster, in only 7 itertions, where as in POLICY ITERATION more number of iterations were required.

- Having said that, VALUE ITERATION does have its draw backs due to its greedy approach.

- In very complex problems, VALUE ITERATIONS may either
  - Not converge to an optimum solution
  - May take even more number of iterations as compared to POLICY iterations

- For simpler problems as the one we see above, VALUE ITERATION can help solve the problem faster.
